# 2.Operation in Jupyternotebook

```python
mix1 = open("1.tab","r")
```

```python
temp = []
for i in mix1:
    temp.append(i.split())
    
mix1.close()
```

List the coverage and position.

```python
coverage = []
position = []
for i in temp:
    coverage.append(i[2])
    position.append(i[1])
    
coverages = []
positions = []
coverages = [ int(x) for x in coverage ]
positions = [ int(x) for x in position ]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(80,20))
plt.plot(positions, coverages,'ro')
plt.ylabel('Depth of Coverage', fontdict={'weight': 'normal', 'size': 50})
plt.xlabel('Position in Genome', fontdict={'weight': 'normal', 'size': 50})
plt.xticks(fontsize=50)
plt.yticks(fontsize=50)
plt.show()
```

```python
print(max(coverages))
print(min(coverages))
print(sum(coverages)/len(coverages))
```

**What is the min, max, and mean coverage across all positions?**

min:1

max:268

mean:37.414

```python
coverages_index = set(coverages)
cou = []
cov = []
for index in coverages_index:
    cou.append(coverages.count(index))
    cov.append(index)
    
```

Calculate the frequency of every coverage.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(80,20))
plt.bar(cov, cou)
plt.xlabel('Depth of Coverage', fontdict={'weight': 'normal', 'size': 50})
plt.ylabel('Count', fontdict={'weight': 'normal', 'size': 50})
plt.xticks(fontsize=50)
plt.yticks(fontsize=50)
plt.show()
```

```python
mix2 = open("2.tab","r")

temp = []
for i in mix2:
    temp.append(i.split())
mix2.close()

coverage2 = []
position2 = []
for i in temp:
    coverage2.append(i[2])
    position2.append(i[1])
```

```python
coverages2 = []
positions2 = []

for a in coverage2:
    coverages2.append(int(a))
for a in position2:
    positions2.append(int(a))

import matplotlib.pyplot as plt
plt.figure(figsize=(80,20))
plt.plot(positions2, coverages2,'ro')
plt.ylabel('Depth of Coverage', fontdict={'weight': 'normal', 'size': 50})
plt.xlabel('Position in Genome', fontdict={'weight': 'normal', 'size': 50})
plt.xticks(fontsize=50)
plt.yticks(fontsize=50)
plt.show()
```

```python
print(max(coverages2))
print(min(coverages2))
print(sum(coverages2)/len(coverages2))
```

```python
coverages2_index = set(coverages2)
cou = []
cov = []
for index in coverages2_index:
    cou.append(coverages2.count(index))
    cov.append(index)
    
import matplotlib.pyplot as plt

plt.figure(figsize=(80,20))
plt.bar(cov, cou)
plt.xlabel('Depth of Coverage', fontdict={'weight': 'normal', 'size': 50})
plt.ylabel('Count', fontdict={'weight': 'normal', 'size': 50})
plt.xticks(fontsize=50)
plt.yticks(fontsize=50)
plt.show()
```
