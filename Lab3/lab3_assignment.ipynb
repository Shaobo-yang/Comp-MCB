### Task 1 Load the script & Produce a dictionary
**1.1** Load the play **_Romeo and Juliet_** into the jupyter notebook as **text**;

**1.2** **Words** is used to split words in the play;

**1.3** **Word_index** is the set of all words.

**1.4** **Counts_dict** is the final dictionary we need.

**1.5** Notation from line 7-line 10 is used to show the elements in dictionary every 2 seconds, just ignore.

**1.6** **Sor** is used to list words by decending order of frequency.  

```python
import string
import time
with open('romeoandjuliet.txt','r') as text:
    words=[raw_word.strip(string.punctuation).lower() for raw_word in text.read().split()]
    words_index=set(words)
    counts_dict={index:words.count(index) for index in words_index}
#for word in sorted(counts_dict,key=lambda x:counts_dict[x],reverse=True):
    #time.sleep(2)
    #print ('{}--{} times'.format(word,counts_dict[word]))
#List sor stored the word that is sorted by their appearance frequency.
sor = sorted(counts_dict,key=lambda x:counts_dict[x],reverse=True)
text.close()
```

### Task 2 Draw the barplot to make the dictionary visiable
**2.1** **Fre** is used to gather the frequency of every word by order of **sor**.

**2.2** It seems that the bar plot below have to be extra large to make the xlabel clear enough to read because my dictionary have about 3700 words, so I just let it go.

```python
fre = []
for i in sor:
    fre.append(counts_dict[i])

import matplotlib.pyplot as plt
import numpy as np
x = np.arange(len(fre))
plt.figure(figsize=(30,10))
plt.xticks(x, sor)
p = plt.bar(range(len(fre)), fre)
plt.show()
```

### Bonus
Draw the scatter plot of log(frequency of words) and log(rank of words)

```python
num = []
ran = []
for i in range(len(fre)):
    num.append(fre[i])
    ran.append(i+1)
import numpy as np
lo_ran = np.log(ran)
lo_num = np.log(num)
plt.figure(figsize=(80,20))
plt.plot(lo_ran, lo_num, 'ro')
plt.ylabel('log[Word count]', fontdict={'weight': 'normal', 'size': 50})
plt.xlabel('log[Rank]', fontdict={'weight': 'normal', 'size': 50})
plt.xticks(fontsize=50)
plt.yticks(fontsize=50)
plt.show()
```
